```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Select only numeric columns from iris
iris_numeric = iris.select_dtypes(include=[np.number])
# Check what I selected
print(iris_numeric.head())

# Calculate correlation matrix on numeric data only
corrmat = iris_numeric.corr()

# Plot heatmap
sns.heatmap(corrmat, annot=True, square=True)
plt.show()
```
