## Exercise 3 - Draw Barchart




```python
import pandas as pd
# Load the data
cars = pd.read_csv('car_data.csv')
# pick the first 10 rows
x = cars['Car_Name'].head(10)
y = cars['Selling_Price'].head(10)
plt.bar(x,y)
# Add labels and title
plt.xlabel('Car Name')
plt.ylabel('Selling Price')
plt.title('First 10 Cars - Selling Price')
# Show the bar chart
plt.show()
```

## Exercise 4 -  Draw Histogram


```python
# Select the 'Selling_Price' column
selling_prices = cars['Selling_Price']
# Create the histogram
plt.hist(selling_prices, bins=10, color='skyblue', edgecolor='black')
# Add labels and title
plt.xlabel('Selling Price (Lakhs)')
plt.ylabel('Number of Cars')
plt.title('Distribution of Selling Prices')
# Show the histogram
plt.show()
```

## Exercise 5 -  Draw Piechart



```python
# Count how many cars of each Fuel Type
fuel_counts = cars['Fuel_Type'].value_counts()
# Create the pie chart
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%') #autopct='%1.1f%%'formats the percentageshown on each slice to 1 decimal place.
# Add title
plt.title('Distribution of Fuel Types')
# Show the pie chart
plt.show()
```
